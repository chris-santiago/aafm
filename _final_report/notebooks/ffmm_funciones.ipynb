```python
import json
import requests
import pandas as pd
import numpy as np
import datetime
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, push_notebook, show
from bokeh.layouts import row
from bokeh.plotting import figure
from ipywidgets import interact
import os
```

```python
def recolector(sy1,sm1,sd1,ey1,em1,ed1):
        #read file with current year data
        data= pd.read_csv(r'C:\Users\bloomberg\Python\FFMM\Fondos_Mutuos_2021_column_ytd.csv')
        data=data.drop(['Unnamed: 0'], axis=1)

        #timeframe you want to include, inputs of the function
        start_date= datetime.datetime(sy1,sm1,sd1)
        end_date=datetime.datetime(ey1, em1,ed1)
        delta=(end_date-start_date).days
        datelist=pd.date_range(start_date, periods=delta).strftime("%Y%m%d").tolist()
        data1=[]
        
        for x in datelist:
            JSONContent = requests.get("https://api.aafm.cl/publicDailyStatistics/" + x).json()
            if 'error' not in JSONContent:
                data1.append(JSONContent)

        for x in range(len(datelist)):
            for y in range(len(data1[x])):
                data1[x][y]["fecha"]=datelist[x]

        subdatos=[]
        for x in range(len(datelist)):
                for y in range(len(data1[x])):
                        subdatos.append([data1[x][y]['administratorName'], 
                             data1[x][y]['fundRUN'], 
                             data1[x][y]['fundName'], 
                             data1[x][y]['fundSeries'], 
                             data1[x][y]['netPatrimony'], 
                             data1[x][y]['efectivePatrimony'], 
                             data1[x][y]['installmentsInCirculation'], 
                             data1[x][y]['providedInstallments'], 
                             data1[x][y]['providedFlow'], 
                             data1[x][y]['rescuedInstallments'],
                             data1[x][y]['rescuedFlow'], 
                             data1[x][y]['installmentValue'],
                             data1[x][y]['totalParticipants'],
                             data1[x][y]['institutionalParticipants'],
                             data1[x][y]['otherParticipants'],
                             data1[x][y]['APV'],
                             data1[x][y]['aafmCategory'],
                             data1[x][y]['svsCategory'],
                             data1[x][y]['svsCategoryId'],
                             data1[x][y]['currency'],
                             data1[x][y]['bloombergCode'],
                             data1[x][y]['pensionFunds'],
                             data1[x][y]['fixedRemuneration'],
                             data1[x][y]['variableRemuneration'],
                             data1[x][y]['affectedExpenses'],
                             data1[x][y]['unaffectedExpenses'],
                             data1[x][y]['investmentCommission'],
                             data1[x][y]['rescuedCommission'],
                             data1[x][y]['fecha']])

        df=pd.DataFrame(subdatos, columns=['administratorName','fundRUN',
         'fundName',
         'fundSeries',
         'netPatrimony',
         'efectivePatrimony',
         'installmentsInCirculation',
         'providedInstallments',
         'providedFlow',
         'rescuedInstallments',
         'rescuedFlow',
         'installmentValue',
         'totalParticipants',
         'institutionalParticipants',
         'otherParticipants',
         'APV',
         'aafmCategory',
         'svsCategory',
         'svsCategoryId',
         'currency',
         'bloombergCode',
         'pensionFunds',
         'fixedRemuneration',
         'variableRemuneration',
         'affectedExpenses',
         'unaffectedExpenses',
         'investmentCommission',
         'rescuedCommission',
         'fecha'])

        data2=data.append(df,ignore_index=True)
        data2.to_csv(r'C:\Users\bloomberg\Python\FFMM\Fondos_Mutuos_2021_column_ytd.csv')
        #in case you are starting a new year
        #df.to_csv(r'C:\Users\bloomberg\Python\FFMM\Fondos_Mutuos_2021_column_ytd.csv')
        #data3= pd.read_csv(r'C:\Users\bloomberg\Python\FFMM\Fondos_Mutuos_2020_column_ytd_1.csv')
        
        #rename and remove existing file, to maintain file name.
        os.remove('C:\Users\bloomberg\Python\FFMM\Fondos_Mutuos_2021_column_ytd.csv')
        os.rename(r'C:\Users\bloomberg\Python\FFMM\Fondos_Mutuos_2021_column_ytd_1.csv',r'C:\Users\bloomberg\Python\FFMM\Fondos_Mutuos_2021_column_ytd.csv')
```
