```python
import pandas as pd
import os
```

```python
# See: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html

# Manually supplying headers because of multi-level headers in oroginal document
headers = ["Period", "Administradora", "Nombre Fondo", "Tipo Fondo", "Moneda", "Serie", "Caract.", "Remuneración: Rem. Fija", "Remuneración: Rem. Var.", "Gastos OP.", "TAC Rem. Fija %", "TAC Rem. Var. %", "TAC Gastos Op. %", "TAC Total %", "Comisión de Colocación: Condición", "Comisión de Colocación: Comisión", "Comisión de Colocación Diferida al Rescate: Condición", "Comisión de Colocación Diferida al Rescate: Comisión"]

data = pd.read_excel('../data/raw/cost/costosffmm20210212.xlsx', sheet_name=0, header=None, names=headers, skiprows=4, parse_dates=["Period"])

data
```

```python
data.dtypes
```

```python
# Drop legend (the last few rows)
data.drop(data.tail(13).index, inplace=True)
data.reset_index(drop=True, inplace=True)

# Remove Empty-ish Rows
data = data[data.Period.notna()]
data = data.reset_index(drop=True)
 
# Convert Types
data["Tipo Fondo"] = data["Tipo Fondo"].astype(int)
 
# TODO: Look at the varieties of strings in this field.
#       Ex: "1.00% anual" and "HASTA 0,5000% ANUAL"
# data["Gastos OP."] = data["Gastos OP."].str.replace("\w%", "").astype(float) / 100

data["TAC Rem. Fija %"] = data["TAC Rem. Fija %"].astype(float) / 100
data["TAC Rem. Var. %"] = data["TAC Rem. Var. %"].astype(float) / 100
data["TAC Gastos Op. %"] = data["TAC Gastos Op. %"].astype(float) / 100
data["TAC Total %"] = data["TAC Total %"].astype(float) / 100

data
```

```python
data.dtypes
```

```python
# Something wonky has happened to the types
# have to re-assign them.
# Some how to related to https://github.com/pandas-dev/pandas/issues/21228

for col_name in data.columns:
    col_type_dirty = data.dtypes[col_name]
    col_type = str if col_type_dirty == "object" else col_type_dirty
    data[col_name] = data[col_name].astype(col_type)
```

```python
output_filepath = '../data/cost.parq'

if os.path.exists(output_filepath):
    os.remove(output_filepath)

data.to_parquet(output_filepath, compression='gzip')
```
