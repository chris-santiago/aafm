```python
#crear nuevo data frame de calculo y hacer id rut+serie
data6= hist3
data6['fecha']=pd.to_datetime(data6['fecha'],format='%Y%m%d')
data6=data6.set_index('fecha')
data6=data6[['administratorName','efectivePatrimony','fundName','installmentValue','pensionFunds','rescuedInstallments','providedInstallments','installmentsInCirculation','svsCategory','fundRUN','fundSeries','aafmCategory','currency']]
fund_run=pd.DataFrame()
fund_run['fundrun_series']=data6['fundRUN']+data6['fundSeries']
data6=data6.drop(columns='fundRUN')
data6=data6.drop(columns='fundSeries')
data6=pd.concat([data6,fund_run],axis=1)

#construir tabla de info fondos
info_adm=pd.DataFrame(data6,columns=['administratorName','fundName','svsCategory','fundrun_series','currency','aafmCategory'] )
info_adm=info_adm.set_index('fundrun_series')

#hacer pivot index fecha columnas idfondos y valores cuotas
data7=data6[['efectivePatrimony','installmentValue','fundrun_series','installmentsInCirculation','aafmCategory']]
pivot_rent=pd.pivot_table(data7, values='installmentValue',columns=['fundrun_series'],index=data7.index)

#hacer pivot index fecha columnas idfondos y valores cuotas,'aafmCategory'
pivot_rent_1=pivot_rent.diff()/pivot_rent

#Serie rentabilidades por serie de tiempo
rent_1d=pivot_rent_1.loc[fecha_actual]
rent_mtd=pivot_rent_1.loc[fecha_mtd:fecha_actual]
rent_ytd=pivot_rent_1.loc[fecha_ytd:fecha_actual]
rent_estadillo=pivot_rent_1.loc[fecha_estallido:fecha_actual]
rent_2019=pivot_rent_1.loc[fecha_2019:fecha_actual]

rent_1d=pd.DataFrame(rent_1d)
rent_mtd=pd.DataFrame(rent_mtd.dropna(axis='columns').sum())*100
rent_ytd=pd.DataFrame(rent_ytd.dropna(axis='columns').sum())*100
rent_estadillo=pd.DataFrame(rent_estadillo.dropna(axis='columns').sum())
rent_2019=pd.DataFrame(rent_2019.dropna(axis='columns').sum())

rent_mtd=rent_mtd.rename(columns={0:"%MTD"})
rent_ytd=rent_ytd.rename(columns={0:"%YTD"})
```

```python
#TABLAS RENTABILIDADES
B_temp=info_adm[['administratorName','fundName','svsCategory','currency','aafmCategory']]
#mtd completo
mtd_all=pd.merge(rent_mtd,B_temp,left_on=rent_mtd.index, right_on=info_adm.index, how="left")
mtd_all.drop_duplicates(subset ="key_0", keep = 'first', inplace = True) 

#ytd completo
ytd_all=pd.merge(rent_ytd,B_temp,left_on=rent_ytd.index, right_on=info_adm.index, how="left")
ytd_all.drop_duplicates(subset ="key_0", keep = 'first', inplace = True) 
```

```python
#Categoria SVS MTD Y YTD
mtd_gby=mtd_all[mtd_all['currency']=='P']
mtd_gby=mtd_gby.groupby('svsCategory').mean()
mtd_gby_pesos=mtd_gby.sort_values(by=['%MTD'],ascending=False)

mtd_gby=mtd_all[mtd_all['currency']=='D']
mtd_gby=mtd_gby.groupby('svsCategory').mean()
mtd_gby_usd=mtd_gby.sort_values(by=['%MTD'],ascending=False)

ytd_gby=ytd_all[ytd_all['currency']=='P']
ytd_gby=ytd_gby.groupby('svsCategory').mean()
ytd_gby_pesos=ytd_gby.sort_values(by=['%YTD'],ascending=False)

ytd_gby=ytd_all[ytd_all['currency']=='D']
ytd_gby=ytd_gby.groupby('svsCategory').mean()
ytd_gby_usd=ytd_gby.sort_values(by=['%YTD'],ascending=False)

total_pesos=pd.merge(mtd_gby_pesos,ytd_gby_pesos,left_on=mtd_gby_pesos.index, right_on=ytd_gby_pesos.index,how="left")
total_pesos=total_pesos.rename(columns={'key_0':"SVS Category"})
total_pesos=total_pesos.style.hide_index()
total_usd=pd.merge(mtd_gby_usd,ytd_gby_usd,left_on=mtd_gby_usd.index, right_on=ytd_gby_usd.index,how="left")
total_usd=total_usd.rename(columns={'key_0':"SVS Category"})
total_usd=total_usd.style.hide_index()
titulo_pesos="Tabla Fondos en Pesos, Categoria SVS"
titulo_dolares="Tabla Fondos en Dolares, Categoria SVS"
display(titulo_pesos,total_pesos,titulo_dolares,total_usd)
```

```python
#Categoria aafm MTD y YTD
mtd_gby=mtd_all[mtd_all['currency']=='P']
mtd_gby=mtd_gby.groupby('aafmCategory').mean()
mtd_gby=mtd_gby.sort_values(by=['%MTD'],ascending=False)

mtd_all1=mtd_all[mtd_all['currency']=='D']
mtd_gby1=mtd_all1.groupby('aafmCategory').mean()
mtd_gby1=mtd_gby1.sort_values(by=['%MTD'],ascending=False)

ytd_gby=ytd_all[ytd_all['currency']=='P']
ytd_gby=ytd_gby.groupby('aafmCategory').mean()
ytd_gby=ytd_gby.sort_values(by=['%YTD'],ascending=False)

ytd_all1=ytd_all[ytd_all['currency']=='D']
ytd_gby1=ytd_all1.groupby('aafmCategory').mean()
ytd_gby1=ytd_gby1.sort_values(by=['%YTD'],ascending=False)

total_pesos1=pd.merge(mtd_gby,ytd_gby,left_on=mtd_gby.index, right_on=ytd_gby.index,how="left")
total_usd1=pd.merge(mtd_gby1,ytd_gby1,left_on=mtd_gby1.index, right_on=ytd_gby1.index,how="left")

total_usd1=total_usd1.rename(columns={'key_0':"AAFM Category"})
total_usd1=total_usd1.style.hide_index()

total_pesos1=total_pesos1.rename(columns={'key_0':"AAFM Category"})
total_pesos1=total_pesos1.style.hide_index()

titulo_pesos1="Tabla Fondos en Pesos, Categoria AAFM"
titulo_usd1="Tabla Fondos en USD, Categoria AAFM"

display(titulo_pesos1,total_pesos1,titulo_usd1,total_usd1)
```

```python
#Mejores y Peores fondos Totales
#SUMAR MAS INFO A LOS FONDOS DE OTRO DATAFRAME
B_temp=info_adm[['administratorName','fundName','svsCategory','currency']]
mtd=pd.merge(rent_mtd,B_temp,left_on=rent_mtd.index, right_on=info_adm.index, how="left")
mtd.drop_duplicates(subset ="key_0", keep = 'first', inplace = True) 

#MEJORES y PEORES FONDOS MTD
mtd_mejores=mtd.sort_values(by=['%MTD'],ascending=False).head(10)
mtd_peores=mtd.sort_values(by=['%MTD'],ascending=True).head(10)
mtd_mejores=mtd_mejores[['%MTD','administratorName','fundName','svsCategory','currency']]
mtd_peores=mtd_peores[['%MTD','administratorName','fundName','svsCategory','currency']]

#SUMAR MAS INFO A LOS FONDOS DE OTRO DATAFRAME
ytd=pd.merge(rent_ytd,B_temp,left_on=rent_ytd.index, right_on=info_adm.index, how="left")
ytd.drop_duplicates(subset ="key_0", keep = 'first', inplace = True) 

#MEJORES y PEORES FONDOS MTD
ytd_mejores=ytd.sort_values(by=['%YTD'],ascending=False).head(10)
ytd_peores=ytd.sort_values(by=['%YTD'],ascending=True).head(10)
ytd_mejores=ytd_mejores[['%YTD','administratorName','fundName','svsCategory','currency']]
ytd_peores=ytd_peores[['%YTD','administratorName','fundName','svsCategory','currency']]

display(mtd_peores,mtd_mejores,ytd_peores,ytd_mejores)
```
